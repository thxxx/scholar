```python
import requests

api_key = ""
url = "https://api.scrapingdog.com/google_scholar"

params = {
    "api_key": "",
    "query": "cnn kim",
    "results": "20",
    "page": 0,
    "language": "en",
    "lr": "lang_en",
    "as_ylo": "2021",
    "as_yhi": "2025",
    "scisbd": True
}

response = requests.get(url, params=params)

if response.status_code == 200:
    data = response.json()
    print(data)
else:
    print(f"Request failed with status code: {response.status_code}")
```
