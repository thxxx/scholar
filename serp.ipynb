```python
import requests
import time
import json
from tqdm import tqdm

names = ['soohyun-park-162375254', 'bobbymin', 'minjaesong', 'ian-ryu', 'chrisheo', 'danielpkl', 'tarsyang', 'dongmin-na-2b92b4253']
linkdapi_apikey = ''

datas = []
for name in tqdm(names):
    time.sleep(1)
    url = f"https://linkdapi.com/api/v1/profile/username-to-urn?username={name}"
    headers = {
        "X-linkdapi-apikey": linkdapi_apikey
    }
    response = requests.get(url, headers=headers)
    print(response.json())
    time.sleep(1)
    urn = response.json().get('data').get('urn')

    url = f"https://linkdapi.com/api/v1/profile/full?username={name}&urn={urn}"
    response = requests.get(url, headers=headers)

    data = response.json()

    single_data = {
        'name': name,
        'urn': urn,
        'data': data
    }
    datas.append(single_data)

    with open('linkdapi_datas.json', 'w') as f:
        json.dump(datas, f, indent=4)
```

```python
min_years = []

for item in data:
    articles = item.get('articles', [])
    if articles:
        years = [a.get('year') for a in articles if a.get('year') is not None]
        if years:
            min_year = min(years)
        else:
            min_year = None
    else:
        min_year = None

    min_years.append(min_year)
```
