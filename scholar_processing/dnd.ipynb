```python
import pandas as pd

df = pd.read_csv("most_recent_total_detail_profiles.csv")
print(len(df))
df.head()
```

```python
import ast 

df['article_list'] = df['articles'].apply(ast.literal_eval)
# df[df['article_list'].apply(len) < 20]
```

```python
def safe_year(year):
    if year == '' or year is None:
        return 9999
    return int(year)

df['the_oldest_article'] = df['article_list'].apply(
    lambda lst: min(lst, key=lambda y: safe_year(y['year']))
)
```

```python
def year_to_bucket(year):
    if year == '':
        return "unknown"
    
    year = int(year)
    if year < 2000:
        return "before 2000"
    elif 2000 <= year <= 2010:
        return "2000 - 2010"
    elif 2010 < year <= 2015:
        return "2010 - 2015"
    elif 2015 < year <= 2019:
        return "2015 - 2019"
    elif 2019 < year <= 2021:
        return "2019 - 2021"
    elif 2021 < year <= 2023:
        return "2021 - 2023"
    else:
        return "2023 - "

from collections import Counter

buckets = df['the_oldest_article'].apply(lambda x: year_to_bucket(x['year']))
cnt = Counter(buckets)
cnt
```

```python
for k, v in cnt.items():
    print(k, v/sum(cnt.values()))
```
